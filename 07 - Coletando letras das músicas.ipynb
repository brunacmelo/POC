```python
list_of_files = sorted( filter( lambda x: os.path.isfile(os.path.join('new_dataset', x)),
                        os.listdir('new_dataset') ) )
```

```python
artists = []
songs_names = []
```

```python
directory = 'new_dataset'
for filename in list_of_files:
    if filename.endswith(".csv"):
        path = directory + '/' + filename
        chart = pd.read_csv(path, encoding='utf-8', delimiter='\t')
        for idx, song_id in enumerate(chart['ID']):
            artists.append(chart.loc[idx, 'artist_names'])
            songs_names.append(chart.loc[idx, 'track_name'])
                    
```
