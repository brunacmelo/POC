```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
# Importing libraries
import sys
from pprint import pprint
import pandas as pd
import os
import json
import numpy as np
import pandas
from lyricsgenius import Genius


from pandas.errors import ParserError
```

```python
list_of_files = sorted( filter( lambda x: os.path.isfile(os.path.join('../data/new_dataset', x)),
                        os.listdir('../data/new_dataset') ) )
```

```python
lyrics_sentiments = pd.read_csv('../data/songs_lyrics_sentiments.csv')
lyrics_sentiments.head()
```

```python
directory = '../data/new_dataset'
for filename in list_of_files:
    if filename.endswith(".csv"):
        path = directory + '/' + filename
        chart = pd.read_csv(path, encoding='utf-8', delimiter='\t')
        compound = np.zeros(chart.shape[0])
        neg = np.zeros(chart.shape[0])
        neu = np.zeros(chart.shape[0])
        pos = np.zeros(chart.shape[0])
        
        for idx, song_id in enumerate(chart['ID']):
            for idx2, song_name in enumerate(lyrics_sentiments['songs_names']):
                chart_name = chart.loc[idx, 'track_name']
                chart_name = chart_name.split(" - ", 1)
                chart_name = chart_name[0]
                if chart_name == song_name:
                    compound[idx] = lyrics_sentiments.loc[idx2, 'compound']
                    neg[idx] = lyrics_sentiments.loc[idx2, 'neg']
                    neu[idx] = lyrics_sentiments.loc[idx2, 'neu']
                    pos[idx] = lyrics_sentiments.loc[idx2, 'pos']
                    
        chart['compound'] = compound
        chart['neg'] = neg
        chart['neu'] = neu
        chart['pos'] = pos
        chart.to_csv(directory + '/' + filename)
            
                    
```

```python
compound_medias = []
neg_medias = []
neu_medias = []
pos_medias = []
```

```python
directory = '../data/new_dataset'
for filename in list_of_files:
    if filename.endswith(".csv"):
        compound_week = []
        neg_week = []
        neu_week = []
        pos_week = []
        path = directory + '/' + filename
        chart = pd.read_csv(path)
        for idx, song_id in enumerate(chart['ID']):
            compound_week.append(chart.loc[idx, 'compound'])
            neg_week.append(chart.loc[idx, 'neg'])
            neu_week.append(chart.loc[idx, 'neu'])
            pos_week.append(chart.loc[idx, 'pos'])
                            
        compound_medias.append(np.mean(compound_week))
        neg_medias.append(np.mean(neg_week))
        neu_medias.append(np.mean(neu_week))
        pos_medias.append(np.mean(pos_week))
```

```python
compound = pd.DataFrame({ 'compound': compound_medias})
neu = pd.DataFrame({ 'neu': neu_medias})
neg = pd.DataFrame({ 'neg': neg_medias})
pos = pd.DataFrame({ 'pos': pos_medias})
```

```python
compound.to_csv('../data/compound.csv')
neu.to_csv('../data/neu.csv')
neg.to_csv('../data/neg.csv')
pos.to_csv('../data/pos.csv')
```
