```python
# Importing required libraries
import requests
import datetime as dt
import os
```

```python
market = 'br' 
period = 'weekly'
delta = 7 
FIRST_DATE = '2018-12-28' # Start date of the first chart collected
YEAR_LIMIT = 2021
```

```python
print('Collecting %s charts...' % market)
start_date = dt.datetime.strptime(FIRST_DATE, "%Y-%m-%d")
end_date = start_date + dt.timedelta(days=delta)

while end_date.year <= YEAR_LIMIT:
    start_date_str = start_date.strftime('%Y-%m-%d')
    end_date_str = end_date.strftime('%Y-%m-%d')
    print('%s | Collecting chart %s -- %s' % (market, start_date_str, end_date_str))

    # Requesting file from Spotify
    url = 'https://spotifycharts.com/regional/%s/%s/%s--%s/download' % (market, period, start_date_str, end_date_str)
    print(url)
    r = requests.get(url)

    # Writing CSV file
    outfile = 'dataset/charts/%s/%d/%s-%s-%s--%s.csv' % (market, end_date.year, market, period, start_date_str, end_date_str) 
    os.makedirs(os.path.dirname(outfile), exist_ok=True)
    with open(outfile, 'w+', encoding="utf-8") as f:
        content = r.text.split('\n', 1)
        f.write(content[1])

    # Update dates
    start_date = end_date
    end_date = start_date + dt.timedelta(days=delta)

print('\n')
print('Finished.')
```
